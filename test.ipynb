```python
# Step 1: Import necessary libraries
import pandas as pd
import json
import matplotlib.pyplot as plt
from IPython.display import display


# Step 2: Load the JSON file
# Replace 'your_json_file.json' with the path to your actual JSON file
json_file_path = 'idea-2024-new.json'

# Open the JSON file and load the data
with open(json_file_path, 'r') as file:
    data = json.load(file)

# Step 3: Convert the JSON data into a DataFrame
df = pd.DataFrame(data)

# Step 4: Explore the DataFrame structure
display(df.head())  # View the first few rows of the DataFrame
print(df.columns)  # View the column names to understand the structure


```

```python
# Group the data by organization and count the number of projects per organization
project_counts = df['organization'].value_counts()  # Count projects per organization

# Count the number of organizations with a given number of projects
org_count_by_project_num = project_counts.value_counts().sort_index()

# Plot the data
plt.figure(figsize=(10,6))
org_count_by_project_num.plot(kind='bar', color='lightblue')
plt.title('Number of Organizations with a Given Number of Projects')
plt.xlabel('Number of Projects')
plt.ylabel('Number of Organizations')
plt.xticks(rotation=0)  # Keep the x-axis labels horizontal for readability

# Add the actual numbers on top of each bar
for i, value in enumerate(org_count_by_project_num):
    plt.text(i, value + 0.1, str(value), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# example of organization with multiple projects submitted

# Count the number of projects per organization
project_counts = df['organization'].value_counts()

# Filter organizations with exactly two projects
org_with_two_projects = project_counts[project_counts == 2]

# Select one example organization
example_organization = org_with_two_projects.index[0]  # Get the name of one organization

# Display the example organization and its projects
example_projects = df[df['organization'] == example_organization]
print(f"Organization: {example_organization}")
print("Projects:")
display(example_projects)
```

```python
# Extract the unique values for "What is the primary issue area that your application will impact?"
unique_issue_areas = df['What is the primary issue area that your application will impact?'].unique()

# Display the unique issue area options
print(unique_issue_areas)
print(len(unique_issue_areas))
```

```python
# Extract 'organization' and 'Direct Impact' columns
# Regular Expression (Regex):

# The regular expression r'(\d+\.?\d*)' is crucial for this operation. Here's how it works:

# \d+: Matches one or more digits (e.g., 2600).
# \.?: Matches an optional decimal point (.). The ? means it might or might not be there.
# \d*: Matches zero or more digits after the decimal point (e.g., 0 in 2,600.0).
# Overall: The regex (\d+\.?\d*) captures both integers (like 500) and floating point numbers (like 2,600.0).
# The () around the expression creates a capturing group, which means the part inside the parentheses is what will be extracted.
df['Direct Impact'] = df['Approximately how many people will be impacted by this project, program, or initiative?'].str.extract(r'(\d+\.?\d*)').astype(float)

# Group by 'Direct Impact' and count the number of organizations for each impact value
impact_counts = df.groupby('Direct Impact')['organization'].nunique().reset_index(name='Organization Count')

# Plot the data
plt.figure(figsize=(10,6))
plt.bar(impact_counts['Direct Impact'], impact_counts['Organization Count'], color='lightgreen')

# Add labels and title
plt.xlabel('Direct Impact (Number of People)')
plt.ylabel('Number of Organizations')
plt.title('Number of Organizations by Direct Impact')
plt.xticks(rotation=45, ha='right')  # Rotate X-axis labels if necessary
plt.tight_layout()

# Step 6: Display the graph
plt.show()
```

```python
impact_column_name = "Approximately how many people will be impacted by this project, program, or initiative?"

# Extract the numeric part using regex and create a new column
df['Direct Impact (Numbers)'] = df[impact_column_name].str.extract(r'(\d+)', expand=False)

# Convert the new column to numeric
df['Direct Impact (Numbers)'] = pd.to_numeric(df['Direct Impact (Numbers)'], errors='coerce')

# Display the updated DataFrame to confirm the new column has been added
display(df[['organization', 'Direct Impact (Numbers)']])

# top 5 highest impact organizations
top_5_organizations = df.sort_values(by='Direct Impact (Numbers)', ascending=False).head(5)

# Display the top 5 organizations
print("top 5 highest direct impact organizations:")
display(top_5_organizations[['organization', 'Direct Impact (Numbers)']])
```

```python
# Count the number of organizations for each goal
goal_counts = df['goal'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))
goal_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Organizations for Each Goal')
plt.xlabel('Goal')
plt.ylabel('Number of Organizations')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the counts on top of the bars
for index, value in enumerate(goal_counts):
    plt.text(index, value, str(value), ha='center', va='bottom')

plt.tight_layout()  # Adjust layout for better fitting
plt.show()
```
